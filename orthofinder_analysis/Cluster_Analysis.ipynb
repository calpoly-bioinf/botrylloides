```python
pathway_dict = {'ko04310': 'WntPath', # Wnts and Fzds
                'ko04350': 'TGFBPath', # TGFB and BMPs and Activin
                'ko04340': 'HedgehogPath', # Hedgehogs
                'ko04330': 'NotchPath', # Notch and Delta
                'ko04012': 'EGFPath', # EGFs
                'ko04668': 'TNFPath', # TNFs
                'ko04060': 'CytokineReceptorIntxnPath', # TNFs, ILs, IFNs, Chemo, Cyto
                'ko00830': 'RetinolMetab'}
                #'ko04710': 'CircadianRhythm'} # Hsa ciracdian rythm
```

```python
polypeptide = []
cluster = []
keggPath = []
pathDescript = []

for row in range(len(clusters_kegg)):
  prot = clusters_kegg.iloc[row]
  pathway = prot.loc['KEGG.pathway']
  for term in pathway_dict:
    if term in pathway:
      polypeptide.append(prot.name[0])
      cluster.append(prot.name[1])
      keggPath.append(term)
      pathDescript.append(pathway_dict.get(term))
  
proteins_paths = pd.DataFrame({'protein': polypeptide, 'cluster': cluster, 'keggPath': keggPath, 'pathDescript': pathDescript})
```

Heatmap Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
proteins_paths_grouped
```

```python
proteins_paths_reset = proteins_paths_grouped.reset_index()
proteins_paths_reset.drop(labels = 'keggPath', axis = 1, inplace = True)
```

```python
proteins_paths_reset['presence'] = 1
```

```python
proteins_paths_reset
```

```python
clusters = np.arange(1, 19)
clusters
```

```python
present_clusters = proteins_paths_reset['cluster'].unique()
present_clusters
```

```python
absent_clusters = [c for c in clusters if c not in present_clusters]
absent_clusters = pd.DataFrame({'pathDescript': 'WntPath', 'protein': np.zeros(len(absent_clusters)), 'cluster': absent_clusters, 'presence': np.zeros(len(absent_clusters))}) # had to put a path so they would show up in pivot table
absent_clusters
```

```python
proteins_paths_reset_concat = pd.concat([proteins_paths_reset, absent_clusters], ignore_index = True)
proteins_paths_reset_concat
```

```python
proteins_paths_reset_concat_indexed = proteins_paths_reset_concat.copy()
proteins_paths_reset_concat_indexed['cluster'] = pd.Categorical(proteins_paths_reset_concat_indexed.cluster, categories=[5, 11, 13, 14, 9, 8, 12, 1, 7, 16, 10, 17, 2, 3, 18, 4, 6, 15], ordered=True)
proteins_paths_reset_concat_indexed.sort_values('cluster', inplace = True)
```

```python
proteins_paths_reset_concat_indexed
```

```python
path_pivot_table_quant = proteins_paths_reset_concat_indexed.pivot_table('protein', columns = 'pathDescript', index = 'cluster', fill_value = 0)
path_pivot_table_quant = path_pivot_table_quant.replace(to_replace = 0, value = np.nan)
path_pivot_table_quant
```

```python
path_pivot_table_presence = proteins_paths_reset_concat_indexed.pivot_table('presence', columns = 'pathDescript', index = 'cluster', fill_value = 0)
path_pivot_table_presence
```

```python
plt.figure(figsize=(8,8))
heatmap_quant = sns.heatmap(path_pivot_table_quant, cmap = "Blues", annot = True, linewidths = 1, linecolor = 'k', vmin = 0, vmax = 30, cbar_kws={"shrink": 0.75, "label": 'Number of Genes'}) # values of np.nan are masked
heatmap_quant.set(xlabel = 'Pathway', ylabel = 'Cluster')
#plt.tick_params(axis = 'x', rotation = 70)
plt.tick_params(axis = 'y', rotation = 0)
for _, spine in heatmap_quant.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(1)
plt.savefig('orthofinder_path_quant_v5.pdf', format = 'pdf', bbox_inches = 'tight', pad_inches = 1)
```
