so i wrote this and now i see that the nomenclature is confusion

orthogroup: a specific orthogroup (OG0000) 

orthogroups: dataframe with the genes per orthogroup per species. use this to convert from a single orthogroup to the genes within them 


```python
def orthogroup2genes(orthogroup, species, file):
    orthogroups = pd.read_table(file)
    ogroup = orthogroups[orthogroups.Orthogroup == orthogroup]
    ogroup_l = list(ogroup[species])
    for i in ogroup_l:
        j = i.split(", ")
    return(j)


def genelist2eggnog(genelist, eggnog_file, outfmt=None):
    eggnog = pd.read_table(eggnog_file, skiprows=4)
    eggnog = eggnog[eggnog['Query'].isin(genelist)]
    eggnog = eggnog[['Query', 'Preferred_name', 'GO terms', 'EC number', 'KEGG KO', 'description']]
    if outfmt is None:
        return eggnog
    elif outfmt == "GO":
        term = "GO terms"
    elif outfmt == "KO":
        term = "KEGG KO"
    elif outfmt == "EC":
        term = "EC number"
    elif outfmt == "description":
        return eggnog[['Query','description']]
    out_terms = eggnog[term].str.split(",").explode().dropna()
    out_terms = list(out_terms)
    return out_terms
    

eggnog_file = "./quest_to_upload/Bviol.eggnog.v2_annotation_04082021_Sheet1.tsv"
orthogroup_file = "./quest_to_upload/proteins/OrthoFinder/Results_May06/Orthogroups/Orthogroups.tsv"
orthocluster_file = "./02_clusters.txt"

def orthogroup2go(orthogroup):
    genelist = orthogroup2genes(orthogroup, 'B_violaceus', orthogroup_file)
    go_list = genelist2eggnog(genelist, eggnog_file, "GO")
    return go_list

def orthogroup2description(orthogroup):
    genelist = orthogroup2genes(orthogroup, 'B_violaceus', orthogroup_file)
    descriptions = genelist2eggnog(genelist, eggnog_file, "description")
    descriptions_list = list(descriptions["description"])
    return descriptions_list

def parse_orthoclusters(orthocluster_file, cluster_num):
    clusters = pd.read_table(orthocluster_file)
    cluster = clusters[clusters["Cluster"] == cluster_num]
    print(list(cluster['Orthogroup']))
    aggregate_terms = []
    aggregate_descriptions = []
    for i in cluster['Orthogroup']:
        tmp_terms = orthogroup2go(i)
        aggregate_terms += tmp_terms
        
        tmp_description = orthogroup2description(i)
        aggregate_descriptions += tmp_description
        
    return (aggregate_terms, aggregate_descriptions)
    
```
